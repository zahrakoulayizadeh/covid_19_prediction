```python
df['date'] = pd.to_datetime(df['date'])  
```

```python
df_2 = df.copy()
df_2
```

```python
df_2.drop(columns = ['iso_code', 'continent', 'location'], inplace=True)
```

```python
train_dataset_country = df_2[df_2['date'] <= '2021-10-31']
```

```python
test_dataset_country = df_2[df_2['date'] <= '2021-11-30']
test_dataset_country = test_dataset_country[test_dataset_country['date'] >= '2021-11-01']
```

```python
def train(model, t):
    y_train = train_dataset_country[t]
    X_train = train_dataset_country.drop(columns = [t, 'date'])
    model.fit(X_train, y_train)
    return model
```

```python
def test(model, t):
    X_test = test_dataset_country.drop(columns = [t, 'date'])
    y_pred = model.predict(X_test)
    y_test = test_dataset_country[t]
    return y_pred, y_test
```

```python
import datetime
base_train_date = datetime.datetime(2021,10, 31)

def predict_days(model, t, model_name):
    days = [7, 14, 21, 28, 30]
    for d in days:
#         print('day=', d)
        y_pred_lst = []
        y_hat_lst = []
        X = df_2[df_2['date'] == base_train_date + datetime.timedelta(days=d)]
        for index, row in X.iterrows():
            y_hat_lst.append(row[t])
            x_t = row.drop([t, 'date'])
            y_pred = model.predict(x_t.values.reshape(1, -1))
            y_pred_lst.append(y_pred)
        plot_graph(d, y_pred_lst, y_hat_lst, t, model_name)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error

metrics_data = []
def metrics(y_test, y_pred, model_name, t):
    mse = mean_squared_error(y_test, y_pred)
    rmse = mse ** (0.5)
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    
    mrae = 0
    for i in range(len(y_test.values)):
        rae = (abs((y_pred[i] - y_test.values[i]) / (y_test.values[i] + 0.00000001))) * 100
        mrae += rae
    
    mrae /= len(y_test.values)
    
    rrse = 0
    for i in range(len(y_test.values)):
        rse = (((y_pred[i] - y_test.values[i]) / (y_test.values[i] + 0.00000001)) ** 2) * 100
        rrse += rse
    
    rrse /= len(y_test.values)
    
    metrics_data.append([model_name, t, mse, rmse, r2, mae, mrae, rrse])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeRegressor

gnb = GaussianNB()
rfr = RandomForestRegressor()
dtr = DecisionTreeRegressor()
```

```python
target = ['new_cases', 'new_deaths', 'total_cases', 'total_deaths']
models = [rfr, dtr, gnb]
for model in models:
    for t in target:
        model = train(model, t)
        y_pred, y_test = test(model, t)
        
        model_name = type(model).__name__
        metrics(y_test, y_pred, model_name, t)

        predict_days(model, t, model_name)
        print(model_name, t)
```

```python
metrics_data_df = pd.DataFrame(data=metrics_data, columns=['model_name', 'target', 'MSE', 'RMSE', 'R2', 'MAE', 'MRAE', 'RRSE'])
metrics_data_df.to_csv('metrics_data_df.csv', index=False)
```
